```python
import pandas as pd
potres=pd.read_csv('potresi.csv')
pd.options.display.max_rows=20
```

```python
#za grafe moramo uporabiti še
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#podatki, ki se jih rabi
#1. vsi podatki združeni skupaj
potres_leto=potres.groupby('leto')
potres_mesec=potres.groupby('mesec')
potres_dan=potres.groupby('dan')
potres_globina=potres.groupby('globina')
potres_magnituda=potres.groupby('magnituda')
#2. povprečja na leto
povp_glo=potres_leto['globina'].mean()
povp_mag=potres_leto['magnituda'].mean()
#3. povprečja na mesec
povp_m_glo=potres_mesec['globina'].mean()
povp_m_mag=potres_mesec['magnituda'].mean()
#4.povprečja na dan
povp_d_glo=potres_dan['globina'].mean()
povp_d_mag=potres_dan['magnituda'].mean()
```
